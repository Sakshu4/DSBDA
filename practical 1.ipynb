```python
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    if pd.isnull(Age):
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 24
        else:
            return Age
```

```python
df['Age']=df[['Age','Pclass']].apply(impute_age,axis=1)
```

```python
df.drop('Cabin',axis=1,inplace=True)
```

```python
df.dropna(inplace=True)
```
